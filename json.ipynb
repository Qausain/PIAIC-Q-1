# **Writting python dictionary in a JSON file**

```
import json
customer_2987 = {"First Son": "faaiz", "Second Daughter": "Anum", "Third": "Sarah"}
with open("customer_2987", "w") as f:
  json.dump(customer_2987, f)
```

# **Reading data from a JSON file**

```
with open("customer_2987.json") as f:
  customer_2987 = json.load(f)
print(customer_2987)
print(customer_2987["Third"])
```
