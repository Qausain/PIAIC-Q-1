# **Required Argument Function**

```
def add_two_numbers(n1, n2):
  return n1 + n2
a = add_two_numbers(2,3)
print(a)
a = add_two_numbers(6,4)
print(a)
a = add_two_numbers(5,9)
print(a)  
```

```
def add_two_numbers(n1, n2):
  print(n1 + n2)

a = add_two_numbers(2, 3)
print(a)  
```

# **Positional Arguments**

```
add_two_numbers(6, 7)
```

```
def subtract_n1_n2(n1, n2):
  return n1 - n2

print(subtract_n1_n2(7, 3))
print(subtract_n1_n2(5, 9))
print(subtract_n1_n2(10, 3))  
```

# **PASSING KEYWORD Argument**

```
def subtract_n1_n2(n1, n2):
  return n1 - n2

print(subtract_n1_n2(n1=7, n2=3))
print(subtract_n1_n2(n2=5, n1=9))
print(subtract_n1_n2(10, 3))    
```

```
def student_profile(name, fname, city, batch, course, time):
  return """
  PIAIC STUDENT CARD
  Name: {}
  Father Name:{}
  City: {}
  Batch: {}
  Course: {}
  Time: {}
  """.format(name, fname, city, batch, course, time)
print(student_profile("Qausain Manzoor", "Manzoor Alam", "Sialkot", 6, "A.I", 10))
```

# **Pass Positional Argument with List**

```
l1 = ["Faaiz", "Qausain", "Sialkot", 6, "A.I", 20]
print(student_profile(*l1))
```

# **Pass Postional Argument with Dictionary**

```
dic = {'name': 'Anum',
       'fname': 'Qausain',
       'city': 'Sialkot',
       'batch': 7,
       'course': 'IOT',
       'time': 4}

print(student_profile(**dic))
```

# **Create function with unlimited parameter**

```
def add_numbers(n1, *l):
  return n1 + sum(l)

add_numbers(2)  
```

```
add_numbers(2, 8)
```

```
add_numbers(32,45,43,43,56,7,7,77,6767,767,6)
```

# **Create function with unlimited parameter (Key words)**

```
def add_key_numbers(n1, **d):
  l1 = []
  l1.append(n1)
  for k, v in d.items():
    l1.append(d[k])
    print(d)
    return sum(l1)

add_key_numbers(n1=7, n2=5, n3=6, n4=7)        
```
