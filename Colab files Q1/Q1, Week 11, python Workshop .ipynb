# **FOR LOOP**

```
for a in [1,2,3,4]:
  print(a)
```

```
for num in range(10):
  print(num, "I am for loop")
```

```
counter = 10
while counter > 0:
  print(counter, "I am while loop")
  counter -= 1
```

```
counter = 0
while counter < 10:
  print(counter, "I am a while loop")
  counter += 1
```

```
fav_nums = []
#fav_nums = 
for num in range(5):
  fav_num = input("Enter your favourite number")
  if fav_num == "q":
    break
  else:
    fav_nums.append(fav_num)
print(fav_nums)
```

```
flag = True
while flag:
  fav_num = input("Enter your favourite number")
  if fav_num == "q":
    flag = False
  else:
    fav_nums.append(fav_num)
print(fav_nums)
```
