```
def power(x, y=2):
  r = 1
  for i in range(y):
    r = r * x
  return r
```

```
power(3)
```

```
power(3, 3)
```

```
def sum(*args):
  '''Function returns the sum
  of all values'''
  r = 0
  for i in args:
    r += i
  return r
```

```
sum(1,2,3)
```

```
sum(1,2,3,4,5)
```
