# **Functions within functions**

```
def comission(sales):
  if sales >100:
    return 100*100
  elif sales>50:
    return 100*50
  elif sales >20:
    return 100* 20
  else:
    return 0      

def salary_calculator(basic, sales):
  gross_salary = basic + comission(sales)
  print(f"Your gross salary is equal to {gross_salary}")

salary_calculator(50000,21)
```
