```
class Car():
    def __init__(self, make, model, year, engine):
        self.make = make
        self.model = model
        self.year = year
        self.engine = engine
        self.odometer_reading = 0   
  
    def get_descriptive_name(self):
        long_name = self.make + " " + self.model + " " + self.engine + " " + str(self.year) 
        return long_name.title()

    def read_odometer(self):
        print("This car has " + str(self.odometer_reading) + " miles on it!") 

    def update_odometer(self, mileage):
        if mileage >= self.odometer_reading:
          self.odometer_reading = mileage
        else:
          print("You can't roll back an odometer!")

    def increment_odometer_reading(self, miles):
        self.odometer_reading += miles
    
    def fill_tank(self):
        print("Petrol tank is Full")
```

```
class Electriccar(Car):
  def __init__(self, make, model, year, engine):
    super().__init__(make, model, year, engine)
    self.battery = Battery("Small", "Exide", "10", "20", "75Amp")

  def fill_tank(self): # overriding function
    print("Electric car doesn't require Petrol")
```

```
class Battery():
  def __init__(self, size, battery_name, battery_weight, battery_plates, battery_amp):
    
    self.battery_size = size     # default attrubute of child class
    self.battery_name = battery_name
    self.battery_weight = battery_weight
    self.battery_plates = battery_plates
    self.battery_amp = battery_amp

  def describe_baterry(self):
    print("This car has a " + str(self.battery_size) + "-Kwh Battery.") 
    print("This car has a " + str(self.battery_name) + " battery.")
    print("This car has a " + str(self.battery_weight) + " Kg.")
    print("This car has a " + str(self.battery_plates) + "Plates.")
    print("This car has a " + str(self.battery_amp) + "-Amp battery.")     
```

```
my_tesla = Electriccar("tesla" , "alpha1" , "2018" , "1800cc")
print(my_tesla.get_descriptive_name())
my_tesla.battery.describe_baterry()
my_tesla.fill_tank()
```
