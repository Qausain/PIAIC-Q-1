```
class User():
  def __init__(self):
    self.name = None
    self.fname = None
    self.password = None
    self.education = None

  def login(name, user_name, password):
    if user_name == "Admin" and password == "admin":
      return "valid user"

  def profile_update(self, name, fname, password, education):
    self.name = name
    self.fname = fname
    self.password = password
    self.education = education

  def view_profile(self):
    print(self.name, self.fname, self.password, self.education)
```

```
u1 = User()
u2 = User()
u3 = User()
```

```
u1.login("Admin","admin")
```

```
u1.view_profile()
```

```
u1.profile_update("ali", "M.Ali", "abc", "MTA")
```

```
u1.view_profile()
```

```
u1.view_profile()
u2.view_profile()
u3.view_profile()
```

# **Inherritance**


```
class Parent():
  def __init__(this, name, fname, course):
    this.name = name
    this.fname = fname
    this.course = course


class Child(Parent):
  
  pass
```

```
p1 = Parent("Qasim", "Aslam", "AI")
```

Multiple Inheritance

```
class Dad():
  def __init__(self):
    self.sound = None
    self.walking_style = None

class Mom():
  def __init__(self):
    super().__init__(Dad)
    self.eye_color = None
    self.hair_color = None
class Child1(Dad, Mom):
  pass
```

```
f1 = Child1()
```

```
class Employee():
    def __init__(self, id, name, salary):
        self.id = id
        self.name = name 
        self.salary = salary

class Developer(Employee):
    def __init__(self,id, name, salary, designation):
        super().__init__(id, name, salary)
        self.designation = designation
        
d1 = Developer(1,"Ali",200000,"Developer")
```
