```
#For Loop
arr = [23, 32, 45, 87, 43]
for elem in arr:
  print(elem)
```

```
#For Loop Break Keyword
arr = [23, 32, 45, 87, 43]
for elem in arr:
  print(elem)
  if elem == 45:
    break
```

```
#For loop continue keyword
arr = [23, 32, 45, 87, 43]
for elem in arr:
  if elem == 45:
    continue
  print(elem)
```

```
#2nd Example of the above code
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    continue
  print(x)
```

```
tablenumber = int(input("Enter Table Number"))
for a in range (1,11):
  print(f"{tablenumber} * {a} = {tablenumber*a}")
```

```
tables = int(input("Enter the range of table"))
for table in range (2, tables+1):
  for num in range(1,11):
    print (f"{table} * {num} = {table*num}")
```

```
# Loop through the tuple. If one of the elements has the same value as my_street, display "yes"
my_street = "Main"
streets = ("Elm", "Walnut", "Center", "Main", "Front")
for a_street in streets:
  if a_street == my_street:
    print("yes")
```

```
outer_loop_total = 0
inner_loop_total = 0
countries = ["Albania", "Morocco", "Brazil", "Denmark"]
capitals = ["Tel Aviv", "Abuja", "Brasília", "Islamabad"]
for country_to_check in countries:
  outer_loop_total += 1
  for city_to_check in capitals:
    inner_loop_total += 1
    if country_to_check == "Brazil" and city_to_check == "Brasília":
      print(outer_loop_total + inner_loop_total)
```

```
# Code nested for loops. When the outer tuple element is 4 and the inner tuple element is "b" display "ok"
nums = (1, 2, 3, 4, 5, 6)
strings = ("a", "b", "c", "d", "e")
for a_number in nums:
  for a_string in strings:
    if a_number == 4 and a_string == "b":
      print("ok")
```
